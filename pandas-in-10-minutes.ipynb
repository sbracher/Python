## Reshaping
https://pandas.pydata.org/pandas-docs/stable/advanced.html#advanced-hierarchical
https://pandas.pydata.org/pandas-docs/stable/reshaping.html#reshaping-stacking

### Stack

```python
tuples = list(zip(*[ ['bar', 'bar', 'baz', 'baz', 'foo', 'foo', 'qux', 'qux'],
                     ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']
                   ]))

index = pd.MultiIndex.from_tuples(tuples, names=['first','second'])

df = pd.DataFrame(np.random.randn(8,2), index=index, columns=['A','B'])
df
```

```python
df2 = df[:4]
df2
```

```python
# The stack() method "compresses" a level in the DataFrame's columns
stacked = df2.stack()
stacked
```

```python
# The unstack() method by default unstacks the last level
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

## Time Series
https://pandas.pydata.org/pandas-docs/stable/timeseries.html#timeseries

```python
# Convert secondly data into 5-minutely data
rng = pd.date_range('1/1/2012', periods=100, freq='S')
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts.resample('5Min').sum()
```

```python
# Time zone representation
rng = pd.date_range('3/6/2012 00:00', periods=5, freq='D')
ts = pd.Series(np.random.randn(len(rng)), rng)
ts
```

```python
ts_utc = ts.tz_localize('UTC')
ts_utc
```

```python
# Convert to another time zone
ts_utc.tz_convert('US/Eastern')
```

```python
# Converting between time span representations
rng = pd.date_range('1/1/2012', periods=5, freq='M')
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

```python
ps = ts.to_period()
ps
```

```python
ps.to_timestamp()
```

```python
# Convert a quarterly frequency with year ending in November to 9am of the end of the month following the quarter end
prng = pd.period_range('1990Q1', '2000Q4', freq='Q-NOV')
ts = pd.Series(np.random.randn(len(prng)), prng)
ts.index = (prng.asfreq('M','e')+1).asfreq('H','s') + 9
ts.head()
```

## Categoricals
https://pandas.pydata.org/pandas-docs/stable/categorical.html#categorical
https://pandas.pydata.org/pandas-docs/stable/api.html#api-categorical

```python
df = pd.DataFrame({ "id" : [1,2,3,4,5,6], 
                    "raw_grade" : ['a', 'b', 'b', 'a', 'a', 'e'] })

df
```

```python
# Convert the raw grades to a categorical data type
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

```python
# Rename the categories to more meaningful names
df["grade"].cat.categories = ["very good", "good", "very bad"]
```

```python
# Reorder the categories and simultaneously add the missing categories
df["grade"] = df["grade"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])
df["grade"]
```

```python
# Sorting is per order in the categories, not lexical order
df.sort_values(by="grade")
```

```python
# Grouping by a categorical column also shows empty categories
df.groupby("grade").size()
```

## Plotting
https://pandas.pydata.org/pandas-docs/stable/visualization.html#visualization

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
# On a DataFrame, the plot() method is a convenience to plot all of the columns with labels
df = pd.DataFrame(np.random.randn(1000, 4), 
                  index=ts.index,
                  columns=['A', 'B', 'C', 'D'])

df = df.cumsum()

plt.figure(); df.plot(); plt.legend(loc='best')
```
